```python
from pandas_datareader import data as web 
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

cotacao_bovespa = yf.download('^BVSP',start="2020-1-1", end = "2021-1-1")
display(cotacao_bovespa)
cotacao_bovespa["Close"].plot(figsize=(15,10))
plt.show

cotacao_petrobras = yf.download('PETR4.SA',start="2020-1-1", end = "2021-1-1")
display(cotacao_petrobras)
cotacao_petrobras["Close"].plot(figsize=(15,10))
plt.show

```
